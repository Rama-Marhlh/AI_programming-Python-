```python
# DEFULT VALYES FOR MUTABLE DATA TYPPS IN THE PEREMETER 
def defualt_with_mutable(mylist=[],num=100):
    mylist.append(num)
    return mylist
#print(defualt_with_mutable())
localo_list=[3,5,7]
print(defualt_with_mutable(localo_list,9))  #[3, 5, 7, 9]
#ret_list=defualt_with_mutable(localo_list)
print(defualt_with_mutable(localo_list))    #[3, 5, 7, 9, 100]
print(defualt_with_mutable(localo_list))    #[3, 5, 7, 9, 100, 100]
print(defualt_with_mutable(localo_list))    #[3, 5, 7, 9, 100, 100]
print(defualt_with_mutable(localo_list,120))#[3, 5, 7, 9, 100, 100, 100, 120]
#print(defualt_with_mutable(ret_list,800))  # try these
#print(defualt_with_mutable([3,5,7],800))

print(defualt_with_mutable())                #[100]
print(defualt_with_mutable())                #[100,100]
print(defualt_with_mutable())                #[100,100,100]
print(defualt_with_mutable(num=99))          #[100,100,100,99]
```

```python
def defualt_with_mutable(mylist=[],num=100):
    mylist.append(num)
    return mylist
local_list=[3,5,7]
print(defualt_with_mutable(local_list,9))  #[3, 5, 7, 9]

ret_list1=defualt_with_mutable(local_list)
print(ret_list1)                           #[3, 5, 7, 9, 100]
print(defualt_with_mutable(local_list))    #[3, 5, 7, 9, 100, 100]
print(ret_list1)                           #[3, 5, 7, 9, 100, 100]
print(local_list)                          #[3, 5, 7, 9, 100, 100]
```

```python
def my_fun(param):
    param.append(4) #append 4 to the list so will be [1, 2, 3, 4]
    param=[4,5,6]   #there is (=) so will be creat new variable differant to mylist in the main
    return param    #[4,5,6]

mylist =[1,2,3]
print("this is my list befor call : ",mylist) #[1, 2, 3]
returend_list=my_fun(mylist)
print("this is my list after call : ",mylist) #[1,2,3,4]
print("this is my returned list from calling the function",returend_list) #[4,5,6]
```

```python
# immutable objects can not be changed 
def fun():
    number=3
    print("the number inside the function is ",number)

number=0
fun()
print("the number in the main is ",number)
```

```python
def fun():
    global number 
    number=3
    print("the number inside the function is ",number)

number=0
fun()
print("the number in the main is ",number)
```
