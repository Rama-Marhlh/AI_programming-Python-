# string Methods

```python
#string methods
#Converts the first character to upper case
p0="welcome to my world"
m0=p0.capitalize()
print(m0)
```

```python
#Returns a centered string
p1="welcome to my world"
m1=p1.center(50)
print(m1)
```

```python
#Returns the number of times a specified value occurs in a string
p2="welcome to my world ,welcome !"
m2=p2.count("welcome")
print(m2)
```

```python
#method returns True if the string ends with the specified value, otherwise False.
p3="welcome to my world."
m3=p3.endswith('.')
print(m3)
```

```python
#startswith() method returns True if the string starts with the specified value, otherwise False.
#method returns True if all the characters are in lower case, otherwise False.
p4="Welcome To my World"
m4=p4.isupper()
print(m4)
#isupper() upper() lower()
```

```python
#method returns True if the string is a valid identifier, otherwise False.
p5="welcome2to2my3world"
m5=p5.isidentifier() #A string is considered a valid identifier if it only contains alphanumeric letters (a-z) and (0-9), or underscores (_). A valid identifier cannot start with a number, or contain any spaces.
print(m5)
```

```python
#Returns True if the string follows the rules of a title
p6="welcome to My World"
m6=p6.title()
print(m6) 
#title() Make the first letter in each word upper case
```

```python
# method removes any (spaces at the beginning)  (spaces at the end)
p7="      hi     "
m7=p7.strip()
print("i wanna say",m7,"to you")
```

```python
#Make the lower case letters upper case and the upper case letters lower case:
p8= "Welcome To My World"
m8=p8.swapcase()
print(m8)   
```

```python
#method splits a string into a list. The splitting is done at line breaks.
p9="Welcome To My World \n welcome"
m9=p9.splitlines()
print(m9)
```

```python
#Check if all the characters in the text are numeric:
p10="1234567"
m10=p10.isnumeric()
print(m10)
```

```python
# method finds the last occurrence of the specified value.
p11="Welcome To My World"
m11=p11.rfind("W")
print(m11)
```

```python
#method returns True if all the characters are alphanumeric
p12="welcome1000"
m12=p12.isalnum()
print(m12)
```

```python
#Check if all the characters in the text are letters:
p13="WelcoToMyWorld"   
m13=p13.isalpha() #(space)!#%&?
print(m13)
```

```python
#Join all items using any character
p14=("welcome","to","my","world")
m14="@".join(p14)
print(m14)
```
